```python
import pandas as pd
from ast import Lambda
import numpy as np
people_1=pd.read_csv("people.csv")
people_1["Firstname"] = people_1["Firstname"].apply(lambda x: x.capitalize().replace(" ", ""))
people_1["Lastname"] = people_1["Lastname"].apply(lambda x: x.capitalize().replace(" ", ""))
people_1["Email"] = people_1["Email"].apply(lambda x: x.replace(" ", ""))
people_1["Phone"]=people_1["Phone"].apply(lambda x: x.replace("-",""))
people_1["Address"] = people_1["Address"].apply(lambda x: x.translate({ord(c): None for c in "No.#"}))

final_people = people_1.drop_duplicates(inplace=False)
final_people.index = np.arange(0, len(final_people))
final_people
```
